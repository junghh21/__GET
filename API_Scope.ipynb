```python
import os
import requests


def check_gemini_api_key_scope(api_key: str):
	url = "https://generativelanguage.googleapis.com/v1beta/models"
	params = {"key": api_key}
	response = requests.get(url, params=params, timeout=10)

	print(f"HTTP {response.status_code}")
	try:
		data = response.json()
	except ValueError:
		print("응답이 JSON이 아닙니다.")
		print(response.text)
		return

	if response.ok:
		print("API 키가 유효하며 Gemini 모델 목록을 조회할 수 있습니다.")
		if "models" in data:
			print("사용 가능한 모델:")
			for model in data["models"]:
				print(f" - {model.get('name')}")
		else:
			print(data)
	else:
		print("API 키 권한 확인 결과:")
		print(data)


def load_env_file(path="./env"):
	if not os.path.isfile(path):
		return
	with open(path, encoding="utf-8") as f:
		for line in f:
			line = line.strip()
			if not line or line.startswith("#") or "=" not in line:
				continue
			key, value = line.split("=", 1)
			key = key.strip()
			value = value.strip().strip('"').strip("'")
			if key and key not in os.environ:
				os.environ[key] = value

load_env_file("./env")

GEMINI_API_KEY = os.environ.get("GEMINI_API_KEY")
if not GEMINI_API_KEY:
	raise RuntimeError("GEMINI_API_KEY 환경변수가 비어있습니다. .env에 설정 후 커널을 재시작하세요.")
check_gemini_api_key_scope(GEMINI_API_KEY)
```
