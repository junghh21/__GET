```python

import requests, json

intraday = "https://www.cboe.com/indices/data/?symbol=VIX&timeline=intraday"
onemonth = "https://www.cboe.com/indices/data/?symbol=VIX&timeline=1M"

response1 = requests.get(onemonth)
response2 = requests.get(intraday)
```

```python
import matplotlib.pyplot as plt

response_dict1 = response1.json()
a = response_dict1['data']
x = [ele[0] for ele in a]
y = [ele[4] for ele in a]
plt.plot(x, y)
plt.show()

response_dict2 = response2.json()
b = response_dict2['data']
x = [ele[0] for ele in b]
y = [ele[4] for ele in b]
plt.plot(x, y)
```
